```python
import matplotlib.pyplot as plt
import numpy as np
import json
import os
import argparse

with open('/Users/max/WASP_AIandML_Assignment/nesy_semanticloss/code/Q1_history/model_epoch_history.json', 'r') as f:
    epoch_history = json.load(f)

with open('/Users/max/WASP_AIandML_Assignment/nesy_semanticloss/code/Q1_history/model_iteration_history.json', 'r') as f:
    iter_history = json.load(f)

print("=== Epoch History     ===")
print(type(epoch_history))
print(epoch_history.keys() if isinstance(epoch_history, dict) else epoch_history[:2])

print("\n=== Iteration History  ===")
print(type(iter_history))
print(iter_history.keys() if isinstance(iter_history, dict) else iter_history[:2])


# Extracting data for plotting
epoch_train_loss = epoch_history['train_loss']
epoch_test_loss = epoch_history['test_loss']
epoch_train_acc = epoch_history['train_acc']
epoch_test_acc = epoch_history['test_acc']
epochs     = range(1, len(epoch_train_loss) + 1)

# Plotting Epoch History
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs, epoch_train_loss, label='Train Loss', marker='o')
plt.plot(epochs, epoch_test_loss, label='Test Loss', marker='o')
plt.title('Loss')
plt.xticks(epochs)  # Set x-ticks to be the epoch numbers
plt.grid(True)  # Add grid for better readability
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(epochs, epoch_train_acc, label='Train Accuracy', marker='o')
plt.plot(epochs, epoch_test_acc, label='Test Accuracy', marker='o')
plt.title('Accuracy')
plt.xticks(epochs)  # Set x-ticks to be the epoch numbers
plt.grid(True)  # Add grid for better readability
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
iter_train_loss = iter_history['train_loss']
iter_test_loss = iter_history['test_loss']
iter_train_acc = iter_history['train_acc']
iter_test_acc = iter_history['test_acc']
iterations = range(1, len(iter_train_loss) + 1)
# Plotting Iteration History
plt.figure(figsize=(20, 5))

plt.plot(iterations, iter_train_loss, label='Train Loss', marker='o')
#plt.plot(iterations, iter_test_loss, label='Test Loss', marker='o')
plt.title('Loss')
batches_per_epoch = 938  # 约938
NUM_EPOCHS = len(epoch_train_loss)  # 从 epoch history 中获取实际的 epoch 数量
plt.xticks(
    [i * batches_per_epoch for i in range(NUM_EPOCHS + 1)],
    [f'Epoch {i}' for i in range(NUM_EPOCHS + 1)]
)
plt.grid(True)  # Add grid for better readability
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.legend()
plt.tight_layout()
```
